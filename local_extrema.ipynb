Per ora data una funzione $f$

```python
import sympy as sp
import numpy as np
# from sympy.printing.latex import LatexPrinter, print_latex

sp.init_printing()
x,y,z = sp.var('x,y,z')
```

```python
f = 3*x**2*y*z - 6*x**2 - 9*y**2*z**2
f
```

```python
gradf = [sp.diff(f,var) for var in [x,y,z]]
eqs = gradf
eqs
```

```python
critics = sp.solve(eqs, [x,y,z], dict=True)

print("Critical points: ")
[p for p in critics]
```

```python
hessian = [sp.diff(gradf[i],var) for i in range(3) for var in [x,y,z]]

second_x = [sp.diff(gradf[0],var) for var in [x,y,z]]
second_y = [sp.diff(gradf[1],var) for var in [x,y,z]]
second_z = [sp.diff(gradf[2],var) for var in [x,y,z]]
# hessian elements:
# hessian[0] = f_xx
# hessian[1] = f_xy
# hessian[2] = f_xz
# hessian[3] = f_yx
# hessian[4] = f_yy
# hessian[5] = f_yz
# hessian[6] = f_zx
# hessian[7] = f_zy
# hessian[8] = f_zz
```

```python
#list containing the hessian matrix calculated at a critical point
hessian_matrices = []

for p in critics:
  f_xx = second_x[0].subs(p)
  f_xy = second_x[1].subs(p)
  f_xz = second_x[2].subs(p)
  f_yx = second_y[0].subs(p)
  f_yy = second_y[1].subs(p)
  f_yz = second_y[2].subs(p)
  f_zx = second_z[0].subs(p)
  f_zy = second_z[1].subs(p)
  f_zz = second_z[2].subs(p)

  hessian_matrices.append(sp.Matrix([[f_xx,f_xy,f_xz],[f_yx,f_yy,f_yz],[f_zx,f_zy,f_zz]]))

hessian_matrices
```

```python
diagonal_hessians = []
for mat in hessian_matrices:
  diagonal_hessians.append(mat.diagonalize())
```
